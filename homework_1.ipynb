```python
import itertools
import random

# Определение стилей
styles = {
    'прическа': ['нет волос', 'длинные в пучок', 'длинные волнистые', 'длинные прямые', 'короткая волнистые', 'короткая прямые', 'короткая курчавые'],
    'цвет волос': ['черный', 'блонд', 'каштановый', 'пастельный розовый', 'рыжий', 'серебристо серый'],
    'аксесуар': ['нет очков', 'круглые очки', 'солнцезащитные очки'],
    'одежда': ['худи', 'комбинезон', 'футболка с круглым вырезом', 'футболка с V-вырезом'],
    'цвет одежды': ['черный', 'синий', 'серый', 'зеленый', 'оранжевый', 'розовый', 'красный', 'белый']
}

# Частоты для каждого атрибута
style_counts = {
    'прическа': [7, 0, 1, 23, 1, 11, 7],
    'цвет волос': [7, 6, 2, 3, 8, 24],
    'аксесуар': [11, 22, 17],
    'одежда': [7, 18, 19, 6],
    'цвет одежды': [4, 5, 6, 8, 6, 8, 7, 6]
}
# 1. Расчет вероятностей для каждого атрибута
style_probabilities = {category: [count / sum(counts) for count in counts] for category, counts in style_counts.items()}

# 2. Создание списка всех возможных комбинаций стилей
all_combinations = list(itertools.product(*styles.values()))

# 3. Расчет общей вероятности для каждой комбинации
combination_probabilities = []
for combination in all_combinations:
    probability = 1.0
    for i, category in enumerate(styles):
        attribute_index = styles[category].index(combination[i])
        probability *= style_probabilities[category][attribute_index]
    combination_probabilities.append(probability)

# 4. Генерация случайного стиля на основе расчитанных вероятностей
random_combination_index = random.choices(range(len(all_combinations)), weights=combination_probabilities, k=1)[0]
random_style = all_combinations[random_combination_index]
random_style_probability = combination_probabilities[random_combination_index]

# Вывод результата
print("Сгенерированный стиль:", dict(zip(styles.keys(), random_style)))
print("Вероятность этого стиля:", random_style_probability)
```

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def collect_pixels(folder='/content/drive/MyDrive/avatars', size=(64, 64)):
    pixels = []
    for filename in os.listdir(folder):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(folder, filename)
            with Image.open(image_path) as img:
                img = img.resize(size).convert('RGB')  # Убедитесь, что изображение в формате RGB
                pixels.append(np.array(img))
    return np.array(pixels)  # Возвращает массив изображений размера [количество_изображений, высота, ширина, каналы]

# Собираем данные пикселей из всех изображений
pixels_data = collect_pixels()

# Размеры изображения
height, width = 64, 64

# Инициализация массивов для хранения распределений
pixel_probabilities = np.zeros((height, width, 3, 256))  # Для каждого пикселя, каждого канала и каждого возможного значения

# Расчет вероятностей
for i in range(height):
    for j in range(width):
        for k in range(3):  # Для каждого канала (R, G, B)
            pixel_values = pixels_data[:, i, j, k]  # Все значения пикселя во всех изображениях
            for value in pixel_values:
                pixel_probabilities[i, j, k, value] += 1
            pixel_probabilities[i, j, k] /= len(pixel_values)  # Нормализация для получения вероятностей

def generate_avatars(probabilities, num_avatars=5):
    new_avatars = []
    for _ in range(num_avatars):
        new_image = np.zeros((height, width, 3), dtype=np.uint8)
        for i in range(height):
            for j in range(width):
                for k in range(3):  # Для каждого канала
                    # Генерация значения пикселя на основе распределения
                    new_pixel_value = np.random.choice(a=256, p=probabilities[i, j, k])
                    new_image[i, j, k] = new_pixel_value
        new_avatars.append(Image.fromarray(new_image))
    return new_avatars

# Генерация и вывод новых аватаров
generated_avatars = generate_avatars(pixel_probabilities)


# Отображение новых аватаров
fig, axes = plt.subplots(1, len(generated_avatars), figsize=(20, 10))
for ax, avatar in zip(axes, generated_avatars):
    ax.imshow(np.asarray(avatar))
    ax.axis('off')  # Не показываем оси
plt.show()

```

```python
from PIL import Image
import numpy as np
import os

def process_images(directory, target_size=(64, 64)):
    images = []
    # Проходим по всем файлам в указанной директории
    for filename in os.listdir(directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Проверка на расширение файла
            img = Image.open(os.path.join(directory, filename)).convert('L')  # Преобразование в оттенки серого
            img = img.resize(target_size, Image.ANTIALIAS)  # Изменение размера изображения
            img_data = np.asarray(img) / 255.0  # Нормализация данных
            images.append(img_data)
    return np.array(images).reshape((-1, 64, 64, 1))  # Добавление размерности канала




train_data = process_images('/content/drive/MyDrive/dataset/train')
proliv_data = process_images('/content/drive/MyDrive/dataset/proliv')


```

```python
import matplotlib.pyplot as plt

# Функция для отображения изображений
def show_images(images, num_images=5):
    fig, axes = plt.subplots(1, num_images, figsize=(20, 20))
    for i, ax in enumerate(axes.flat):
        if i < len(images):
            ax.imshow(images[i].reshape(64, 64), cmap='gray')
            ax.axis('off')
    plt.show()

# Вывод первых пяти изображений из train_data
print("Train data samples:")
show_images(train_data, 5)

# Вывод первых пяти изображений из proliv_data
print("Proliv data samples:")
show_images(proliv_data, 5)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D

input_img = Input(shape=(64, 64, 1))

# Шаг 2: Создание автоэнкодера
input_img = Input(shape=(64, 64, 1))
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

x = Conv2D(32, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

autoencoder.compile(optimizer='adam', loss='mean_squared_error')
autoencoder.fit(train_data, train_data, epochs=50, batch_size=128, shuffle=True)
# Сохранение модели
autoencoder.save('/content/drive/MyDrive/my_autoencoder_model.h5')

```

```python
from sklearn.metrics import mean_squared_error

def compute_mse(model, data):
    predicted = model.predict(data)
    mse = mean_squared_error(data.reshape((data.shape[0], -1)), predicted.reshape((predicted.shape[0], -1)), multioutput='raw_values')
    return mse

train_mse = compute_mse(autoencoder, train_data)
proliv_mse = compute_mse(autoencoder, proliv_data)
threshold = np.quantile(train_mse, 0.90)
```

```python
print("Минимальное значение MSE для train-данных:", min(train_mse))
print("Максимальное значение MSE для train-данных:", max(train_mse))
```
